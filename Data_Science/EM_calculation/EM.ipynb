```python
import numpy as np
```

```python
# ** E step **
p = 0.1
q = 0.9

for index in range(10) :
    print("\n\n [ ",index+1,"th loop ]")
    
    # for example
#     nH = [5,9,8,4,7]
#     nT = [5,1,2,6,3]
    nH = [5,9,8]
    nT = [5,1,2]
    totalAH = []
    totalAT = []
    totalBH = []
    totalBT = []
    
    for i in range(len(nH)) :
        print("\n\n [ ",i+1,"th experiment ]")
        P_SA = np.power(p,nH[i])*np.power(1-p,nT[i])
        P_SB = np.power(q,nH[i])*np.power(1-q,nT[i])
        #print("P_SA : %.2f"%P_SA, ", P_SB : %.2f"% P_SB)

        P_AS = round(P_SA/(P_SA+P_SB),2)
        P_BS = 1 - P_AS
        print("P_AS : %.2f"%P_AS, ", P_BS : %.2f"%P_BS)

        AH = round(nH[i] * P_AS,2)
        AT = round(nT[i] * P_AS,2)
        BH = round(nH[i] * P_BS,2)
        BT = round(nT[i] * P_BS,2)
        
        print("AH : %.2f"%AH, ", AT : %.2f"%AT)
        print("BH : %.2f"%BH, ", BT : %.2f"%BT)

        totalAH.append(AH)
        totalAT.append(AT)
        totalBH.append(BH)
        totalBT.append(BT)
    
    totalAH = np.sum(totalAH)
    totalAT = np.sum(totalAT)
    totalBH = np.sum(totalBH)
    totalBT = np.sum(totalBT)

    print("\n*** Total Result ***")
    print("A : %.2f"%totalAH,"H, %.2f"%totalAT, "T")
    print("B : %.2f"%totalBH,"H, %.2f"%totalBT, "T")

    # ** M step **
    p = totalAH / (totalAH + totalAT)
    q = totalBH / (totalBH + totalBT)
    print("p : %.2f"%p, ", q : %.2f"%q)
    
    if index == 10 :
        break
```
